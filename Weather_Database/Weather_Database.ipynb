```python
# Importing dependencies

import pandas as pd
import numpy as np
from citipy import citipy
import time
from datetime import datetime
import requests
# importing sys to grab weather_api_key that isn't in weather_database folder
import sys

#moving path up a level to World_Weather_Analysis folder
sys.path.insert(1, '..')
from config import weather_api_key
```

```python
# Creating set of 2000 random latitutde and longitude combinations, putting it in a list

lats = np.random.uniform(low=-90.000, high=90.000, size=2000)
lngs = np.random.uniform(low=-180.000, high=180.000, size=2000)
lat_lngs = list(zip(lats, lngs))
coordinates = lat_lngs
```

```python
# Creating empty list to hold cities

cities = []
```

```python
# Finding the nearest city for reach latitude and longitutde combination using for loop

for coordinate in coordinates:
    city = citipy.nearest_city(coordinate[0], coordinate[1]).city_name
    
    #Adding city to list if the city is unique
    
    if city not in cities:
        cities.append(city)
        
#checking length of list
len(cities)
```

```python
#creating list to hold weather data
city_data = []
```

```python
#Creating URL for Weather Map API Call

url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + weather_api_key
print(url)
```

```python
# Grabbing weather information

# Print the beginning of the logging

print("Beginning Data Retrieval     ")
print("-----------------------------")

# initializing counters
record_count = 1
set_count = 1

# looping through all cities
for i, city in enumerate(cities):
    
    # Grouping cities by 50 to avoid overloading weather API
    
    if(i % 50 == 0 and i >= 50):
        set_count += 1
        record_count = 1
        time.sleep(60)
        
    #Create endpoint URL with each city
    city_url = url + "&q=" + city.replace(" ","+")
    
    #logging URL, record, set numbers, and city name
    
    print(f"Processing Record {record_count} of Set {set_count} | {city}")
    
    # Add 1 to the record count
    record_count += 1
    
# Run an API request for each of the cities.
    try:
        #Parse the JSON and retreive Data
        city_weather = requests.get(city_url).json()
        
        #parsing data
        city_lat = city_weather["coord"]["lat"]
        city_lng = city_weather["coord"]["lon"]
        city_max_temp = city_weather["main"]["temp_max"]
        city_humidity = city_weather["main"]["humidity"]
        city_clouds = city_weather["clouds"]["all"]
        city_wind = city_weather["wind"]["speed"]
        city_desc = city_weather["weather"][0]["description"]
        
        #Append city information into city_data list
        city_data.append({"City": city.title(),
                          "Lat": city_lat,
                          "Lng": city_lng,
                          "Max Temp": city_max_temp,
                          "Humidity": city_humidity,
                          "Cloudiness": city_clouds,
                          "Wind Speed": city_wind,
                          "Current Description": city_desc})
    except:
        print("City not found. Skipping...")
        pass
    
# Indicate that Data Loading is complete.
print("-----------------------------")
print("Data Retrieval Complete      ")
print("-----------------------------")   
```

```python
#converting city data to dataframe
city_data_df = pd.DataFrame(city_data)
```
